```python
image=cv2.imread("/content/Fig0304(a)(breast_digital_Xray).tif")
plt.imshow(image,cmap="gray")
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cols,rows=gray_image.shape
image2=np.zeros((cols,rows),dtype=np.uint8)
image_thrash=np.zeros((cols,rows),dtype=np.uint8)
for i in range(cols):
  for j in range(rows):
    neg_image=255-gray_image[i,j]
    image2[i,j]=neg_image
    if image2[i,j]<124.5:
      image_thrash[i,j]=0
    else:
      image_thrash[i,j]=255
fig=plt.figure()
plt.subplot(2,2,1)
plt.title("Gray")
plt.imshow(gray_image,cmap="gray")
plt.subplot(2,2,2)
plt.title("Negative")
plt.imshow(image2,cmap="gray")
```
